```python
#4.7
inauguration_date = pd.to_datetime('2021-01-20')

#Creating a column to indicate pre or post inauguration
dataframe['Period'] = dataframe['DATE OCC'].apply(lambda x: 'Pre-Biden' if x < inauguration_date else 'Post-Biden')

crime_counts = dataframe.groupby([dataframe['DATE OCC'], 'Period']).size().reset_index(name='Crime_Count')

#Calculating 30-day moving average
crime_counts['Moving_Average'] = crime_counts.groupby('Period')['Crime_Count'].rolling(window=30).mean().reset_index(0, drop=True)

plt.figure(figsize=(12, 6))
for period in ['Pre-Biden', 'Post-Biden']:
    data = crime_counts[crime_counts['Period'] == period]
    plt.plot(data['DATE OCC'], data['Moving_Average'], label=period)

plt.axvline(x=inauguration_date, color='r', linestyle='--', label='Inauguration')
plt.title('Crime Rate Before and After Biden\'s Inauguration')
plt.xlabel('Date')
plt.ylabel('30-Day Moving Average of Crime Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculating average crime rates for each period
average_rates = crime_counts.groupby('Period')['Crime_Count'].mean()
print("Average Daily Crime Rates:")
print(average_rates)
```

```python
#Performing t-test to check if the difference is statistically significant
from scipy import stats

pre_biden = crime_counts[crime_counts['Period'] == 'Pre-Biden']['Crime_Count']
post_biden = crime_counts[crime_counts['Period'] == 'Post-Biden']['Crime_Count']

t_stat, p_value = stats.ttest_ind(pre_biden, post_biden)
print(f"\nt-statistic: {t_stat}")
print(f"p-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in crime rates between the two periods.")
else:
    print("There is no significant difference in crime rates between the two periods.")
```

as the p value is less than alpha value (alpha = 0.05) there is a significant difference in crime rates between the two periods.
